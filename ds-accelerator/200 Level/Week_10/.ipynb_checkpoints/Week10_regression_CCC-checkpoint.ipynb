```python
bigmartclean = bigmart.copy()
bigmartclean_target = bigmartclean["Item_Outlet_Sales"]
bigmartclean_features =bigmartclean.drop('Item_Outlet_Sales', axis=1)
```

```python
# TODO Split train and test (test_size=0.2, ran5dom_state=42)

from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(bigmartclean_features, bigmartclean_target, test_size=0.25, random_state=42)
```

```python
# TODO
#clearn null values from Item_Weight
print(bigmartclean["Item_Weight"].describe())
pivot_table = bigmartclean.pivot_table(values='Item_Weight', index='Item_Fat_Content', aggfunc='mean')
print(pivot_table)
```

```python
avg_weight = features_train["Item_Weight"].mean()

features_train["Item_Weight"]= features_train["Item_Weight"].fillna(avg_weight)
features_test["Item_Weight"]= features_test["Item_Weight"].fillna(avg_weight)
```

```python
#scaling data

#scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(features_train)
features_train_scaled = scaler.transform(features_train)
features_test_scaled = scaler.transform(features_test)

print("Training data shape fitted:", features_train_scaled.shape)
print("Testing data shape fitted:", features_test_scaled.shape)
```

#### Linear Regression

Fit a linear model

```python
# TODO
from sklearn.linear_model import LinearRegression

reg = LinearRegression()
```

Look at estimated intercept and coefficients

```python
# Todo
reg.fit(features_train_scaled, target_train)
print(reg.score(features_train_scaled, target_train))
print(reg.coef_)
print(reg.intercept_)


coef_df = pd.DataFrame({"Feature": features_train.columns, "Coefficient": reg.coef_}, 
                       columns=["Feature", "Coefficient"])

coef_df.sort_values("Coefficient", ascending = 0)
```

Make predictions on test dataset and plot the true outcome compared to the predicted outcome

```python
# Todo
y_pred_train =reg.predict(features_train_scaled)
y_pred = reg.predict(features_test_scaled)
plt.scatter(y_pred, target_test)
plt.title('y_pred vs target')
plt.xlabel('y_pred' )
plt.ylabel('target')
```

Calculate the residual sum of squares and mean squared error

```python
#residual Analysis 
plt.scatter(y_pred_train, y_pred_train- target_train, c="blue", marker ='o', label = 'training data')
plt.scatter(y_pred, y_pred - target_test, c="orange", marker ='*', label = 'test data')
plt.xlabel('Predicted' )
plt.ylabel('Residuals')
plt.legend(loc='upper left')
plt.hlines(y=0, xmin=-10, xmax= 50, color = 'k')
plt.show()


#MSE: 
from sklearn.metrics import mean_squared_error
print("Train MSE: ", mean_squared_error(target_train, y_pred_train))
print("Test MSE: ", mean_squared_error(target_test, y_pred))


#Coef of Determiniation
from sklearn.metrics import r2_score
print("Train r2_score: ", r2_score(target_train, y_pred_train))
print("Test r2_score: ", r2_score(target_test, y_pred))




#Adjusted R2
#Train
r2_train = reg.score(features_train_scaled, target_train)
n_train =features_train_scaled.shape[0]
p_train = features_train_scaled.shape[1]
Adjusted_r2_train = 1-(1-r2_train)*(n_train-1)/(n_train-p_train-1)
print("Adjusted_r2_train", Adjusted_r2_train)


#Test
r2_test = reg.score(features_test_scaled, target_test)
n_test =features_test_scaled.shape[0]
p_test = features_test_scaled.shape[1]
Adjusted_r2_test = 1-(1-r2_test)*(n_test-1)/(n_test-p_test-1)
print("Adjusted_r2_test", Adjusted_r2_test)
```

We've learned that there are some caveats in the train/test split method. In order to avoid this, we can perform something called cross validation. Try K-Fold Cross Validation (ex. 5-fold)

```python
# Todo Cross Val Score
from sklearn.model_selection import cross_val_score
scores = cross_val_score(reg, features_train_scaled, target_train, cv =5)
print(scores)

print("Accuracy Mean:", scores.mean(),"Accuracy Std:", scores.std())
```

#### Regularized Linear Regression

If you look at the coefficients, they are very large in magnitude which signifies overfitting. To cater to this, lets use a ridge regression model.

```python
# Todo

from sklearn.linear_model import Ridge

ridge_mod = Ridge(alpha=.5, normalize =True)
ridge_mod.fit(features_train_scaled, target_train)
print(ridge_mod.score(features_train_scaled, target_train))
print(ridge_mod.coef_)
print(ridge_mod.intercept_)

y_pred_train_ridge =ridge_mod.predict(features_train_scaled)
y_pred_ridge = ridge_mod.predict(features_test_scaled)
plt.scatter(y_pred_ridge, target_test)
plt.title('y_pred vs target')
plt.xlabel('y_pred' )
plt.ylabel('target')
plt.show()

coef_df_ridge = pd.DataFrame({"Feature": features_train.columns, "Coefficient": ridge_mod.coef_}, 
                       columns=["Feature", "Coefficient"])

print(coef_df_ridge.sort_values("Coefficient", ascending = 0))

#Coef of Determiniation
from sklearn.metrics import r2_score
print("Train r2_score: ", r2_score(target_train, y_pred_train_ridge))
print("Test r2_score: ", r2_score(target_test, y_pred_ridge))
```

There are so many predictors in the regression model. Can you use a lasso model to apply feature selection?

```python
# Todo
# Todo

from sklearn.linear_model import Lasso

lasso_mod = Lasso(alpha=.5, normalize =True)
lasso_mod.fit(features_train_scaled, target_train)
print(lasso_mod.score(features_train_scaled, target_train))
print(lasso_mod.coef_)
print(lasso_mod.intercept_)

y_pred_train_lasso =lasso_mod.predict(features_train_scaled)
y_pred_lasso = lasso_mod.predict(features_test_scaled)
plt.scatter(y_pred_lasso, target_test)
plt.title('y_pred vs target')
plt.xlabel('y_pred' )
plt.ylabel('target')
plt.show()

coef_df_lasso = pd.DataFrame({"Feature": features_train.columns, "Coefficient": lasso_mod.coef_}, 
                       columns=["Feature", "Coefficient"])

print(coef_df_lasso.sort_values("Coefficient", ascending = 0))

#Coef of Determiniation
from sklearn.metrics import r2_score
print("Train r2_score: ", r2_score(target_train, y_pred_train_lasso))
print("Test r2_score: ", r2_score(target_test, y_pred_lasso))
```

What are some differences between ridge and lasso, when should we proceed to use each?

```python
# Todo
#Ridge Regression: Ridge regression does not perform feature selection, as it retains all features and only reduces their impact.
#Lasso Regression: Lasso regression can perform feature selection by setting the coefficients of irrelevant features to zero. It automatically selects the most relevant features and discards the rest.
```
