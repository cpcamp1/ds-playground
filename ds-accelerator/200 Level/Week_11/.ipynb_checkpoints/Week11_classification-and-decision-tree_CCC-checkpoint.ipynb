# Week11 Classification & Decision Tree

We continued learning other algorithms of Supervised Learning. Classification belongs to the category of supervised learning where the targets also provided with the input data. 

Classification is the process of predicting the class of given data points. Classes are sometimes called as targets/ labels or categories. Classification predictive modeling is the task of approximating a mapping function (f) from input variables (X) to discrete output variables (y).

A decision tree is a largely used non-parametric effective machine learning modeling technique for regression and classification problems. To find solutions a decision tree makes sequential, hierarchical decision about the outcomes variable based on the predictor data.

Seperate the dataset as response variable and feature variabes

```python
# TODO

target = wine["quality"]
features= wine.drop("quality", axis = 1)
```

Split to training set and testing set (Note: test_size = 0.25, random_state = 42)

```python
# TODO
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features, target, test_size=0.25, random_state=42)
```

Apply Standard scaling to get optimized result, think of why we need to do this?

```python
# TODO
#scaling data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(features_train)
features_train_scaled = scaler.transform(features_train)
features_test_scaled = scaler.transform(features_test)

print("Training data shape fitted:", features_train_scaled.shape)
print("Testing data shape fitted:", features_test_scaled.shape)
```

### Modeling

#### Logistic Regression

Train a Binary Classifier and fit with training set

```python
# TODO
from sklearn.linear_model import LogisticRegression

Log = LogisticRegression()
Log.fit(features_train_scaled, target_train)

print("Coefficients:", Log.coef_)
print("Intercept:", Log.intercept_)

#coef_df = pd.DataFrame({"Feature": features_train.columns, "Coefficient": Log.coef_}, columns=["Feature", "Coefficient"])
```

Perform predictions on test data

```python
# Todo
y_pred = Log.predict(features_test_scaled)

accuracy = Log.score(features_test_scaled, target_test)
print(accuracy)

```

We've learned that there are some caveats in the train/test split method. In order to avoid this, we can perform something called cross validation. Try K-Fold Cross Validation (ex. 10-fold)

```python
# Todo
# Todo Cross Val Score
from sklearn.model_selection import cross_val_score
scores = cross_val_score(Log, features_train_scaled, target_train, cv =10)
print(scores)

print("Accuracy Mean:", scores.mean(),"Accuracy Std:", scores.std())
```

Print out Confusion Matrix and explain the Four Quadrants

```python
# Todo
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix

y_pred_cross = cross_val_predict(Log, features_train_scaled, target_train, cv =10)

confusion_matrix(target_train, y_pred_cross)

#Q1: Top Left: Pred neg; actual neg
#Q2: Top Right: pred pos, actual neg
#Q3: bottom Left pred neg; actual Pos
#Q4: Bottom Right: Pred pos; actual pos

```

What's AUC for your Logistic Regression model?

```python
# Todo
from sklearn.metrics import precision_score, recall_score, f1_score

precision_score =precision_score(target_train, y_pred_cross)
print("precision_score:", precision_score)


recall_score =recall_score(target_train, y_pred_cross)

print("recall_score:", recall_score)


f1_score = f1_score(target_train, y_pred_cross)
print("f1_score:", f1_score)
```

#### Regularized Logistic Regression

Build a Binary Classifier with a L2 Regularizer and print out its Confusion Matrix

```python
# NOTE: For both questions that follow, use solver = 'liblinear' 
# More on why here at the docs: https://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LogisticRegression.html
```

```python
# Todo
Ridge = LogisticRegression(penalty='l2', solver='liblinear')

Ridge = Ridge.fit(features_train_scaled, target_train)
y_pred_train_ridge = Ridge.predict(features_train_scaled)

confusion_matrix(target_train, y_pred_train_ridge)
```

Build a Binary Classifier with a L1 Regularizer and print out its Confusion Matrix

```python
# Todo
Lasso = LogisticRegression(penalty='l1', solver='liblinear')

Lasso = Lasso.fit(features_train_scaled, target_train)
y_pred_train_Lasso =Lasso.predict(features_train_scaled)

confusion_matrix_lasso = confusion_matrix(target_train, y_pred_train_Lasso)
confusion_matrix_lasso
```

#### SGD

Fit a SGD model and perform predictions on test set

```python
# Todo
from sklearn.linear_model import SGDClassifier
from sklearn.metrics import accuracy_score

SGD = SGDClassifier(loss='log', random_state=42)

SGD.fit(features_train_scaled, target_train)

y_pred_SDG_train = SGD.predict(features_train_scaled)
y_pred_SDG_test = SGD.predict(features_test_scaled)

accuracy_train = accuracy_score(target_train, y_pred_SDG_train)
accuracy_test = accuracy_score(target_test, y_pred_SDG_test)
print("Accuracy Train:", accuracy_train)
print("Accuracy Test:", accuracy_test)

confusion_matrix_SGD_train = confusion_matrix(target_train, y_pred_SDG_train)
confusion_matrix_SGD_Test = confusion_matrix(target_test, y_pred_SDG_test)
print(confusion_matrix_SGD_train)
print(confusion_matrix_SGD_Test)
```

#### Decision Tree

Build a Decision Tree Classifier and perform predictions on test set

```python
from sklearn import tree
tree_model1 = tree.DecisionTreeClassifier(max_depth=7)
tree_model2 = tree.DecisionTreeClassifier(max_depth=3)

tree_model1.fit(features_train_scaled, target_train)
tree_model2.fit(features_train_scaled, target_train)

y_pred1 = tree_model1.predict(features_train_scaled)
y_pred2 = tree_model2.predict(features_train_scaled)

y_predtest1 = tree_model1.predict(features_test_scaled)


confusion_matrix_tree_train1 = confusion_matrix(target_train, y_pred1)
confusion_matrix_tree_train2 = confusion_matrix(target_train, y_pred2)
confusion_matrix_tree_test1 = confusion_matrix(target_test, y_predtest1)
print("confusion_matrix_tree_train1:", confusion_matrix_tree_train1)
print("confusion_matrix_tree_train2:", confusion_matrix_tree_train2)
print("confusion_matrix_tree_test1:", confusion_matrix_tree_test1)

precision_train  = confusion_matrix_tree_train1[1, 1] / (confusion_matrix_tree_train1[1, 1] + confusion_matrix_tree_train1[0, 1])
print(precision_train)

precision_test = confusion_matrix_tree_test1[1, 1] / (confusion_matrix_tree_test1[1, 1] + confusion_matrix_tree_test1[0, 1])
print(precision_test)


#recall_score =recall_score(target_train, y_pred1)
#f1_score = f1_score(target_train, y_pred1)
```

```python
# TODO
from sklearn import tree


results_tree = pd.DataFrame(columns=['Max_depth', 'Confusion Matrix Train', 'Confusion Matrix Test'
                                     ,'Precision Score Train','Precision Score Test'
                                , 'Recall Score', 'F1 Score'])
max_depth = [3, 4, 5, 6, 7, 8, 9, 10]
for n in max_depth:
    tree_model1 = tree.DecisionTreeClassifier(max_depth=n)
    tree_model1.fit(features_train_scaled, target_train)
    y_pred = tree_model1.predict(features_train_scaled)
    y_predtest1 = tree_model1.predict(features_test_scaled)
    cmtrain = confusion_matrix(target_train, y_pred)
    cmtest = confusion_matrix(target_test, y_predtest1)
    precision_train=  cmtrain[1, 1] / (cmtrain[1, 1] + cmtrain[0, 1])
    precision_test =  cmtest[1, 1] / (cmtest[1, 1] + cmtest[0, 1])
    recall = cmtest[1, 1] / (cmtest[1, 1] + cmtest[1, 0])
    f1 = 2 * (precision_test * recall) / (precision_test + recall)
    results_tree = results_tree.append({'Max_depth': n, 'Confusion Matrix Train': cmtrain,'Confusion Matrix Test': cmtest 
                                        ,'Precision Score Train': precision_train
                                        ,'Precision Score Test': precision_test 
                                        ,'Recall Score': recall, 'F1 Score': f1
                                       },
                                   ignore_index=True)
results_tree

```
